```python
import random

class Sentiment:
    NEGATIVE = "NEGATIVE"
    NEUTRAL = "NEUTRAL"
    POSITIVE = "POSITIVE"
class Review:
    def __init__(self,text,score):
        self.text=text
        self.score=score
        self.sentiment=self.get_sentiment()
        
    #to get the sentiment
    def get_sentiment(self):
        if self.score <= 2:
            return Sentiment.NEGATIVE
        elif self.score == 3:
            return Sentiment.NEUTRAL
        else:
            return Sentiment.POSITIVE #4 or 5
            
class ReviewContainer:
    def __init__(self, reviews):
        self.reviews = reviews
        
    def get_text(self):
        return [x.text for x in self.reviews]#for splitting traing
    
    def get_sentiment(self):
        return [x.sentiment for x in self.reviews]#for getting test dta

    def evenly_distribute(self):
        negative = list(filter(lambda x: x.sentiment == Sentiment.NEGATIVE, self.reviews))
        positive = list(filter(lambda x: x.sentiment == Sentiment.POSITIVE, self.reviews))
        positive_shrunk = positive[:len(negative)]
        self.reviews = negative + positive_shrunk
        random.shuffle(self.reviews)

        
file_name = "/drive/data/ml_scikit/Books_small.json"

reviews = []

with open(file_name) as f:
    for line in f:
        review = json.loads(line)
        reviews.append(
            (Review(review["reviewText"], review["overall"])) 
        )

        
print("Number of reviews:", len(reviews))

print("6th review:")
print(reviews[5])
print(reviews[5].get_sentiment())

print("Review text:")
#print(reviews[5][0])
print(reviews[5].text)
print("Rating:")
#print(reviews[5][1])
print(reviews[5].score)
```

```python
import sklearn
from sklearn.model_selection import train_test_split

training, test = train_test_split(reviews, test_size=0.33, random_state=42)
#train_x =(x for x in training.text())

train_container = ReviewContainer(training)#class ReviewContainer: 
test_container = ReviewContainer(test)

train_container.evenly_distribute()
train_x = train_container.get_text()
train_y = train_container.get_sentiment()

test_container.evenly_distribute()#to balance the number of positive and negative reviews
test_x = test_container.get_text()
test_y = test_container.get_sentiment()

print(train_y.count(Sentiment.POSITIVE))
print(train_y.count(Sentiment.NEGATIVE))
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# This book is great !
# This book was so bad
print(type(train_x[0]))
print(train_x[0])
vectorizer = TfidfVectorizer()
train_x_vectors = vectorizer.fit_transform(train_x)
print(train_x_vectors)#train_x contains Review objects, not text.

test_x_vectors = vectorizer.transform(test_x)
print(test_x_vectors)
print(train_x[0])#This book is great!
print(train_x_vectors[0].toarray())#to get a regular NumPy array: 

```

CLASSIFICATIONS

```python
#Linear SVM
from sklearn import svm

clf_svm = svm.SVC(kernel='linear')

clf_svm.fit(train_x_vectors, train_y)

test_x[0]

clf_svm.predict(test_x_vectors[0])
#array(['POSITIVE'], dtype='<U8')Scikit-learn returns predictions as a NumPy array.

"""
fruits = ["apple", "banana", "orange"]
print(fruits[0])
Output:
apple
"""
prediction = clf_svm.predict(test_x_vectors[0])[0]
print(prediction)
```

```python
#decision trees
from sklearn.tree import DecisionTreeClassifier

clf_dec = DecisionTreeClassifier()
clf_dec.fit(train_x_vectors, train_y)

clf_dec.predict(test_x_vectors[0])
#array(['NEGATIVE'], dtype='<U8')
```

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB

clf_gnb = DecisionTreeClassifier()
clf_gnb.fit(train_x_vectors, train_y)

clf_gnb.predict(test_x_vectors[0])
#array(['NEGATIVE'], dtype='<U8')
```

```python

#Logistic Regression
from sklearn.linear_model import LogisticRegression

clf_log = LogisticRegression()
clf_log.fit(train_x_vectors, train_y)
clf_log.predict(test_x_vectors[0])
```

```python
print(clf_svm.score(test_x_vectors, test_y))#to see which ine performed better
print(clf_dec.score(test_x_vectors, test_y))
print(clf_gnb.score(test_x_vectors, test_y))
print(clf_log.score(test_x_vectors, test_y))
```

```python
#test data
test_set = ['very fun', "bad book do not buy", 'horrible waste of time']
new_test = vectorizer.transform(test_set)

clf_svm.predict(new_test)
print(clf_svm.predict(new_test[0:1])[0])
print(clf_svm.predict(new_test[2])[0])
print(clf_svm.predict(new_test[2:3])[0])

#more better printing
predictions = clf_svm.predict(new_test)
for review, pred in zip(test_set, predictions):
    print(f"{review} --> {pred}")
```

```python
#for fine tuning and finding the best parameters for the prediction
#Grid Search + Cross Validation
from sklearn.model_selection import GridSearchCV

parameters = {
    'C': [0.1, 1, 5, 10],
    'kernel': ['linear', 'rbf'] #will ttain 5*2 models
}

grid = GridSearchCV(svm.SVC(), parameters, cv=5)#cv=5 suppose you have 100 reviews Scikit-learn splits data into 5 parts. 

grid.fit(train_x_vectors, train_y)

print(grid.best_params_)#{'C': 5, 'kernel': 'linear'}
print(grid.best_score_)# average cross-validation accuracy
print(grid.score(test_x_vectors, test_y))
#list1 =[]
#print(list1.append(clf.cv_results_)) It does not return the updated list.It returns:None
#print(list1)
```
